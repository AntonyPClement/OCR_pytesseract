```python
def maxmin(image):
    return (np.max(image)-image)
```

```python
from PIL import Image
```

```python
ip=Image.open(r"D:\road-sign-2.jpg")
gray=ip.convert('L')
```

```python
bw=np.asarray(gray).copy()

bw[bw<64]=0
bw[bw>=64]=255

plot.imshow(bw)
```

```python
x=pytesseract.image_to_string(bw, timeout=10)
print(x)
```

```python
from gtts import gTTS
```

```python
language = 'en'
myobj = gTTS(text=x, lang=language, slow=False) 
```

```python
myobj.save("ocr1.mp3")
```
